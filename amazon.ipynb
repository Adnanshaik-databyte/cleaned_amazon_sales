renamingh the columns for more readability

```python
df.rename(columns={
    "product_name": "ProductName",
    "discounted_price": "DiscountedPrice",
    "actual_price": "ActualPrice",
    "discount_percentage": "DiscountPercent",
    "rating": "Rating",
    "rating_count": "RatingCount",
    "MainCategory": "Category",
    "discount_amount": "DiscountAmount",
    "product_link": "ProductLink",
    "img_link": "ImageLink"
}, inplace=True)
```

```python
df.info()
```

"ratingcount" has just one null value
Remove commas and convert to numeric

```python

df['RatingCount'] = df['RatingCount'].replace(',', '', regex=True).astype(float)

df['RatingCount'] = df['RatingCount'].fillna(df['RatingCount'].mean())
```

```python
df.head(4)
```

```python
df["MainCategory"] = df["Category"].apply(lambda x: x.split('|')[0] if isinstance(x, str) else x)
```

```python
df['MainCategory'].head(100)
```

```python
df[['DiscountedPrice','ActualPrice','DiscountPercent']]
```

```python
def clean_currency(val):
    if isinstance(val, str):
        return float(val.replace('₹', '').replace(',', '').strip())
    return val

def clean_percentage(val):
    if isinstance(val, str):
        return float(val.replace('%', '').strip())
    return val

def clean_number(val):
    if isinstance(val, str):
        val = val.replace(',', '').strip()
        return int(val) if val.isdigit() else None
    return val
```

```python
df["DiscountedPrice"] = df["DiscountedPrice"].apply(clean_currency)
df["ActualPrice"] = df["ActualPrice"].apply(clean_currency)
df["DiscountPercent"] = df["DiscountPercent"].apply(clean_percentage)
df["Rating"] = pd.to_numeric(df["Rating"], errors='coerce')
df["RatingCount"] = df["RatingCount"].apply(clean_number)
```

```python
df.head(5)
```

```python
df["DiscountAmount"] = df["ActualPrice"] - df["DiscountedPrice"]
```
